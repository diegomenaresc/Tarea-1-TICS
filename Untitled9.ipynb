***1.1 Estudien el contexto del que surge el dataset, revisando su descripción y propósito.***

Contiene información desde el momento en que el cliente realiza un pedido hasta la entrega y evaluación final del producto, incluyendo detalles de pago, tiempos logísticos, ubicación geográfica de vendedores y clientes, y reseñas de satisfacción.

*   La relación entre **categorías de producto, métodos de pago y regiones geográficas.**
*   El **rendimiento de los vendedores** y su impacto en la experiencia del usuario.

En el contexto académico, este dataset es especialmente útil para practicar el **ciclo completo de analítica de datos**, desde la **modelación relacional** (creación del esquema y normalización) hasta el **procesamiento y análisis predictivo**, permitiendo formular y resolver problemas reales de aprendizaje automático relacionados con comercio electrónico.

```python
from graphviz import Digraph
from IPython.display import Image, display

g = Digraph('ER_Olist', format='png')
g.attr(rankdir='LR', splines='ortho', nodesep='0.6', ranksep='1', fontsize='10')

def entity(name, pk, attrs, color="#f6f8fa"):
    # Construye una “tabla” HTML para el nodo
    rows = [f'<TR><TD BGCOLOR="#e6f0ff"><B>{name}</B></TD></TR>']
    rows.append(f'<TR><TD align="left"><B>PK</B>: {pk}</TD></TR>')
    if attrs:
        rows.append(f'<TR><TD align="left">{attrs}</TD></TR>')
    label = f'<<TABLE BORDER="1" CELLBORDER="0" CELLSPACING="0" BGCOLOR="{color}">' + "".join(rows) + '</TABLE>>'
    g.node(name, label=label, shape='plaintext')

# Entidades
entity("CUSTOMER", "customer_id",
       "customer_unique_id<br/>customer_zip_code_prefix<br/>customer_city<br/>customer_state")

entity("SELLER", "seller_id",
       "seller_zip_code_prefix<br/>seller_city<br/>seller_state")

entity("ORDER", "order_id",
       "customer_id (FK)<br/>order_status<br/>order_purchase_timestamp<br/>order_approved_at<br/>order_delivered_carrier_date<br/>order_delivered_customer_date<br/>order_estimated_delivery_date<br/><I>order_city_snapshot (violación 2FN si estuviera en ORDER_ITEM)</I>")

entity("PRODUCT", "product_id",
       "product_category_name<br/>product_name_length<br/>product_description_length<br/>product_photos_qty<br/>product_weight_g<br/>product_length_cm<br/>product_height_cm<br/>product_width_cm<br/><I>product_category_department (violación 3FN)</I>")

entity("ORDER_ITEM", "order_id + order_item_id",
       "product_id (FK)<br/>seller_id (FK)<br/>shipping_limit_date<br/>price<br/>freight_value<br/><I>order_city_snapshot (violación 2FN)</I>")

entity("ORDER_PAYMENT", "order_id + payment_sequential",
       "payment_type<br/>payment_installments<br/>payment_value")

entity("ORDER_REVIEW", "review_id",
       "order_id (FK)<br/>review_score<br/>review_comment_title<br/>review_comment_message<br/>review_creation_date<br/>review_answer_timestamp")

entity("CATEGORY", "category_name", "category_translation")

entity("GEOLOCATION", "zip_prefix + lat + lng",
       "geolocation_city<br/>geolocation_state")

# Relaciones (dirección meramente estética)
g.edge("CUSTOMER", "ORDER", label="1..N")
g.edge("ORDER", "ORDER_ITEM", label="1..N")
g.edge("PRODUCT", "ORDER_ITEM", label="1..N")
g.edge("SELLER", "ORDER_ITEM", label="1..N")
g.edge("ORDER", "ORDER_PAYMENT", label="1..N")
g.edge("ORDER", "ORDER_REVIEW", label="0..1")
g.edge("CATEGORY", "PRODUCT", label="1..N")
g.edge("GEOLOCATION", "CUSTOMER", label="ref ZIP")
g.edge("GEOLOCATION", "SELLER", label="ref ZIP")

out_path = g.render('/content/er_olist_graphviz', cleanup=True)
display(Image(filename=out_path))
print("Diagrama generado en:", out_path)
```

***1.2 Esquema entidad–relación (ER) con violaciones intencionales a 2FN y 3FN***

**a) Entidades, atributos y llaves primarias (ER inicial)**

Este ER inicial incluye una violación a 2FN y una a 3FN, tal como exige el enunciado. En el punto de normalización, ambas se corrigen hasta 3FN.

**CUSTOMER**

* **PK:** customer_id
* Atributos: customer_unique_id, customer_zip_code_prefix, customer_city, customer_state

**SELLER**

* **PK:** seller_id
* Atributos: seller_zip_code_prefix, seller_city, seller_state

**ORDER**

* **PK:** order_id
* **FK:** customer_id → CUSTOMER.customer_id
* Atributos: order_status, order_purchase_timestamp, order_approved_at, order_delivered_carrier_date, order_delivered_customer_date, order_estimated_delivery_date

**ORDER_ITEM** (violación 2FN en este ER inicial)

* **PK compuesta:** (order_id, order_item_id)
* **FK:** order_id → ORDER.order_id, product_id → PRODUCT.product_id, seller_id → SELLER.seller_id
* Atributos: shipping_limit_date, price, freight_value, order_city_snapshot ⟵ (depende solo de order_id)

**PRODUCT** (violación 3FN en este ER inicial)
* **PK:** product_id
* Atributos: product_category_name, product_name_length, product_description_length, product_photos_qty, product_weight_g, product_length_cm, product_height_cm, product_width_cm, product_category_department ⟵ (depende de product_category_name)

**ORDER_PAYMENT**
* **PK compuesta:** (order_id, payment_sequential)
* **FK:** order_id → ORDER.order_id
* Atributos: payment_type, payment_installments, payment_value

**ORDER_REVIEW**
* **PK:** review_id
* **FK:** order_id → ORDER.order_id
* Atributos: review_score, review_comment_title, review_comment_message, review_creation_date, review_answer_timestamp

**CATEGORY** (opcional, si se usa el CSV de traducciones/jerarquías)
* **PK:** category_name
* Atributos: category_translation

**GEOLOCATION**
* **PK compuesta:** (geolocation_zip_code_prefix, geolocation_lat, geolocation_lng)
* Atributos: geolocation_city, geolocation_state

***1.3 Relaciones y cardinalidades***

* **CUSTOMER (1) — (N) ORDER** (un cliente puede tener muchos pedidos)
* **ORDER (1) — (N) ORDER_ITEM** (un pedido contiene varios ítems)
* **PRODUCT (1) — (N) ORDER_ITEM** (un producto aparece en muchos ítems)
* **SELLER (1) — (N) ORDER_ITEM** (un vendedor provee muchos ítems)
* **ORDER (1) — (N) ORDER_PAYMENT** (un pedido puede tener múltiples pagos)
* **ORDER (1) — (0..1) ORDER_REVIEW** (un pedido puede tener a lo más una reseña)
* **CATEGORY (1) — (N) PRODUCT** (si se factoriza categoría)
* **GEOLOCATION }o—o{ CUSTOMER** y **GEOLOCATION }o—o{ SELLER** (asociación por prefijo ZIP)

***1.4 Violaciones de normalización***

* **Violación 2FN (en ORDER_ITEM):**
PK = (order_id, order_item_id). El atributo order_city_snapshot **no depende de toda la PK**, sino solo de order_id (y, por transitividad, del customer_id del pedido). Esto es **dependencia parcial** ⇒ viola **2FN.**

Corrección (sección de normalización): mover order_city_snapshot a ORDER.

* **Violación 3FN (en PRODUCT):**
product_category_department depende de product_category_name (atributo **no clave**). Es una **dependencia transitiva** respecto a la PK product_id ⇒ viola **3FN.**

Corrección (sección de normalización): factorizar CATEGORY(category_name PK, category_department) y referenciar desde PRODUCT por FK category_name.

```python
from graphviz import Digraph
from IPython.display import Image, display

g3 = Digraph('ER_Olist_3FN', format='png')
g3.attr(rankdir='LR', splines='ortho', nodesep='0.6', ranksep='1', fontsize='10')

def entity(name, pk, attrs, color="#f6f8fa"):
    rows = [f'<TR><TD BGCOLOR="#e6f0ff"><B>{name}</B></TD></TR>']
    rows.append(f'<TR><TD align="left"><B>PK</B>: {pk}</TD></TR>')
    if attrs:
        rows.append(f'<TR><TD align="left">{attrs}</TD></TR>')
    label = f'<<TABLE BORDER="1" CELLBORDER="0" CELLSPACING="0" BGCOLOR="{color}">' + "".join(rows) + '</TABLE>>'
    g3.node(name, label=label, shape='plaintext')

# Entidades normalizadas
entity("CUSTOMER", "customer_id",
       "customer_unique_id<br/>customer_zip_code_prefix<br/>customer_city<br/>customer_state")

entity("SELLER", "seller_id",
       "seller_zip_code_prefix<br/>seller_city<br/>seller_state")

entity("ORDER", "order_id",
       "customer_id (FK)<br/>order_status<br/>order_purchase_timestamp<br/>order_approved_at<br/>order_delivered_carrier_date<br/>order_delivered_customer_date<br/>order_estimated_delivery_date<br/>order_city_snapshot")

entity("CATEGORY", "category_name", "category_department")

entity("PRODUCT", "product_id",
       "category_name (FK)<br/>product_name_length<br/>product_description_length<br/>product_photos_qty<br/>product_weight_g<br/>product_length_cm<br/>product_height_cm<br/>product_width_cm")

entity("ORDER_ITEM", "order_id + order_item_id",
       "product_id (FK)<br/>seller_id (FK)<br/>shipping_limit_date<br/>price<br/>freight_value")

entity("ORDER_PAYMENT", "order_id + payment_sequential",
       "payment_type<br/>payment_installments<br/>payment_value")

entity("ORDER_REVIEW", "review_id",
       "order_id (FK)<br/>review_score<br/>review_comment_title<br/>review_comment_message<br/>review_creation_date<br/>review_answer_timestamp")

entity("GEOLOCATION", "zip_prefix + lat + lng",
       "geolocation_city<br/>geolocation_state")

# Relaciones
g3.edge("CUSTOMER", "ORDER", label="1..N")
g3.edge("ORDER", "ORDER_ITEM", label="1..N")
g3.edge("PRODUCT", "ORDER_ITEM", label="1..N")
g3.edge("SELLER", "ORDER_ITEM", label="1..N")
g3.edge("ORDER", "ORDER_PAYMENT", label="1..N")
g3.edge("ORDER", "ORDER_REVIEW", label="0..1")
g3.edge("CATEGORY", "PRODUCT", label="1..N")
g3.edge("GEOLOCATION", "CUSTOMER", label="ref ZIP")
g3.edge("GEOLOCATION", "SELLER", label="ref ZIP")

out_path3 = g3.render('/content/er_olist_3FN_graphviz', cleanup=True)
display(Image(filename=out_path3))
print("Diagrama final 3FN generado en:", out_path3)
```
